```python
import matplotlib.pyplot as plt

# Data
x_values = ['A', 'B', 'C', 'D', 'E']
y_values = [50, 24, 36, 40, 15]

# Create a bar chart
plt.bar(x_values, y_values)

# Add labels and title
plt.xlabel('Goods')
plt.ylabel('Quantity sold in millions')
plt.title('Bar Graph of Goods Sold')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 5, 6, 7]

# Create a line chart
plt.plot(x_values, y_values)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Example Line Graph')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def my_func(x):
    return x**2 + 3*x - 2

# Define the x range for the plot
x_range = np.linspace(-5, 5, 100)

# Calculate the y values for the curve
y_values = my_func(x_range)

# Create the curve graph
plt.plot(x_range, y_values)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Example Curve Graph')

# Display the graph
plt.show()
```
